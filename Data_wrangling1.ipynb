```python
# Importing required libraries

import pandas as pd
import numpy as np
```

```python
# Loading dataset

df = pd.read_csv("Titanic-Dataset.csv")

# Display first 5 rows
df.head()
```

```python
# Checking dimensions

df.shape
```

```python
# Checking data types


df.dtypes
```

```python
# Type conversion

df['Survived'] = df['Survived'].astype('category')

df.dtypes
```

```python
# Normalizing Age column

df['Age_Normalized'] = (
    (df['Age'] - df['Age'].min()) /
    (df['Age'].max() - df['Age'].min())
)

df[['Age', 'Age_Normalized']].head()
```

```python
# Convert categorical variable into quantitative variable

df['Sex'] = df['Sex'].map({
    'male': 0,
    'female': 1
})

df.head()
```
