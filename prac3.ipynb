```python
import pandas as pd
import numpy as np
```

```python
data = {
    'Name': ['A', 'B', 'C', 'D', 'E', 'F'],
    'Age_Group': ['Young', 'Young', 'Adult', 'Adult', 'Senior', 'Senior'],
    'Income': [25000, 30000, 40000, 45000, 35000, 30000]
}
```

```python
df1 = pd.DataFrame(data)
```

```python
print("Dataset:\n")
print(df1)
```

```python
group = df1.groupby('Age_Group')
```

```python
print("\nSummary Statistics:\n")
print(group['Income'].agg(['mean', 'median', 'min', 'max', 'std']))
```

```python
print("\nIncome List:\n")
print(group['Income'].apply(list))
```
